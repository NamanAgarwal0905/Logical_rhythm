**PREPROCESSING AND INSIGHTTS ON THE ALGORITHM THAT SHOULD BE USED**


*   Initially i felt the dataset doesnt require much preprocessing only label encoding followed by the splitting of engine column into parts would be enough and as it turns out my minimal approach was enough to get me the top spot on the leaderboard :)

*   the main thing about the data is its non linear relation with pricing


*   Since we are using label encoding its virtually impossible to fit a straight line or even a curved line for that matter because the dependency on features like the model or brand or paint is quite uneven and spiky as visible in graphs above

*   therefore I ruled out the linear and polynomial regression because linear progression would just perform way too poorly while my intution says that polynomial regression would somehow overfit on the train set considering the large degree of plynomial which will be required to fit the spiky data

*   Therefore i decided to go ahead with random forest regression which is quite good for scattered data as it divides the planes into parts based on the decision tree mechanism








---



```python
df['clean_title'] = df['clean_title'].fillna('No')
df['fuel_type'] = df['fuel_type'].fillna('Any')
test_df['clean_title'] = test_df['clean_title'].fillna('No')
test_df['fuel_type'] = test_df['fuel_type'].fillna('Any')

combined_df = pd.concat([df, test_df], axis=0)
categorical_cols = ['brand', 'model', 'fuel_type', 'transmission', 'ext_col', 'int_col', 'accident', 'clean_title','engine']

label_encoders = {}
for col in categorical_cols:
    le = LabelEncoder()
    combined_df[col] = le.fit_transform(combined_df[col].astype(str))
    label_encoders[col] = le

# Split combined data back into train and test
train_encoded = combined_df.iloc[:len(df)]
test_encoded = combined_df.iloc[len(df):]

```

```python
test_encoded
```

```python
X = train_encoded.drop(columns=['id', 'target'])
y = train_encoded['target']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=42)

rf = RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)

y_pred = rf.predict(X_test)

mae = mean_absolute_error(y_test, y_pred)
rmse = mean_squared_error(y_test, y_pred, squared=False)

print("Mean Absolute Error:", mae)
print("Root Mean Squared Error:", rmse)
```

```python
X_test = test_encoded.drop(columns=['id','target'])
predictions = rf.predict(X_test)

output = pd.DataFrame({'id': test_df['id'], 'target': predictions})
output.to_csv('predictions.csv', index=False)

print("Predictions saved to predictions.csv")
```

```python
import re
```

```python

def extract_engine_features(engine_str):
    hp_match = re.search(r'(\d+\.?\d*)HP', engine_str)
    volume_match = re.search(r'(\d+\.?\d*)L', engine_str)

    horsepower = float(hp_match.group(1)) if hp_match else None
    engine_volume = float(volume_match.group(1)) if volume_match else None

    return horsepower, engine_volume

df[['engine_hp', 'engine_volume']] = df['engine'].apply(lambda x: pd.Series(extract_engine_features(x)))
```

```python
df[['engine_hp', 'engine_volume']] = df[['engine_hp', 'engine_volume']].fillna('0')
```

**MOJOR ISSUE FACED**


*   the only issue i faced while making the model was label encoding
*   so some of the data that is models or different fuel types or engine present in the train datset were not a part of the test datset
*   for that reason the model was misinterpreting certain things for example if gasoline was encoded as 3 in the train data it got the number 2 in test data due to some fuel types missing








---



**SOLUTION**


*   one solution was to do tne label encoding manually which obviously wasnt possible
*   I came up with a trick . I combined the test and train dataset together for sometime and then lable encoded the combined dataset which made sure that there is no difference in the encodings between the two sets after which i seperated them .This was a very useful trick and saved me time





```python
train_encoded
```

**TRAINING THE MODEL**


*   For the training i had a hyper parameter called the n_estimators which basically means the number of tress in the random forst (as we know random forest is an ensemble learning algorithm based on decision trees)
*   Initially I tested with 100 trees which did'nt give that good of a result so i decided to increase the number of trees to 500 as there were a hell lot of parameters to handle which gave much better results at least good enough for the leaderboard



```python
X = train_encoded.drop(columns=['id', 'target','engine'])
y = train_encoded['target']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.05, random_state=69)

rf = RandomForestRegressor(n_estimators=500, n_jobs=-1, random_state=69)
rf.fit(X_train, y_train)

y_pred = rf.predict(X_test)

mae = mean_absolute_error(y_test, y_pred)
rmse = mean_squared_error(y_test, y_pred, squared=False)

print("Mean Absolute Error:", mae)
print("Root Mean Squared Error:", rmse)
```

SAVING THE MODEL

```python
X_test = test_encoded.drop(columns=['id','target','engine'])
predictions = rf.predict(X_test)

output = pd.DataFrame({'id': test_df['id'], 'target': predictions})

output.to_csv('predictions3.csv', index=False)

print("Predictions saved to predictions3.csv")
```
